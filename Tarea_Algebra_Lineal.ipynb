####2. Implemente el algoritmo de Gauss-Seidel para resolver el sistema de ecuaciones visto en clase.


```python
def GaussSeidel(A,b,x0, itmax = 100, tolerancia = 1e-16):

  x = x0.copy()

  for it in range(itmax):

    x_new = np.copy(x)

    for i in range(A.shape[0]):
      Suma = 0.
      for j in range(A.shape[1]):
        if i != j:
          Suma += A[i,j]*x_new[j]
          x_new[i] = (b[i] - Suma)/A[i,i]
    if np.linalg.norm( np.dot(A,x_new)-b) < tolerancia:
      break

    x = np.copy(x_new)
  return x, it

A = np.array([[3,-1,-1],[-1,3,1],[2,1,4]])
b = np.array([1.,3.,7.])

x0 = np.array([0.,0.,0.])
x,_ = GaussSeidel(A,b,x0)
print(x)
```

####3. En Python, implemente una clase para solucionar sistemas lineales.  El constructor de clase deber recibir la matriz y el vector independiente.El primer método de clase resuelve el sistema usando el método de Jacobi, y el segundo método resuelve el sistema usando el método de Gaus-Seidel. En la pantalla debe aparecer el número de iteraciones necesarias para resolver el problema.


```python
class LinearSolver:
    def __init__(self, A, b, tol=1e-16, max_iter=1000):

        self.A = np.array(A)
        self.b = np.array(b)
        self.tol = tol
        self.max_iter = max_iter

    def Jacobi(self):
      A = np.copy(self.A)
      b = np.copy(self.b)
      x = np.zeros_like(b)

      for it in range(self.max_iter):

        x_new = np.copy(x)

        for i in range(A.shape[0]):

          Suma = 0.
          for j in range(A.shape[1]):
            if i != j:
              Suma += A[i,j]*x[j]
              x_new[i] = (b[i] - Suma)/A[i,i]

    # Criterio de convergencia
        if np.linalg.norm( np.dot(A,x_new)-b) < self.tol:
          break

        x = np.copy(x_new)

      return x, it

    def GaussSeidel(self):

      A = np.copy(self.A)
      b = np.copy(self.b)
      x = np.zeros_like(b)

      for it in range(self.max_iter):

        x_new = np.copy(x)

        for i in range(A.shape[0]):

          Suma = 0.
          for j in range(A.shape[1]):
            if i != j:
              Suma += A[i,j]*x_new[j]
              x_new[i] = (b[i] - Suma)/A[i,i]

    # Criterio de convergencia
        if np.linalg.norm( np.dot(A,x_new)-b) < self.tol:
          break

        x = np.copy(x_new)

      return x, it
```

####4. Implemente un algoritmo que realiza la multiplicación de dos matrices. Use el algoritmo para calcular:

####8. Implemente la descomposición LU para factorizar la siguiente matriz A=LU

```python
def descomposicion(A):
    n = A.shape[0]
    L = np.eye(n)  #matriz identidad
    U = A.copy()

    for i in range(n):
        for j in range(i+1, n):

            a = U[j, i] / U[i, i]
            L[j, i] = a
            U[j, :] -= a * U[i, :]

    return L, U
```

####12. Usando los métodos de Newton Rapson y descenso del gradiente, encuentre la solución de los siguientes sistemas de ecuaciones no lineales:

a) Use x(0)= (2,2)

```python
G1=(lambda x,y: np.log((x**2) + (y**2)) - np.sin(x * y) - np.log(2) - np.log(np.pi),\
    lambda x,y: np.exp(x - y) + np.cos(x * y))

def GetF2(G,r):

  n = r.shape[0]
  v = np.zeros(n)

  for i in range(n):
    v[i] = G[i](r[0],r[1])

  return v

x0 = np.array([2,2])

def GetJacobian2(f,r,h=1e-3):

  n = r.shape[0]

  J = np.zeros((n,n))

  for i in range(n):
    for j in range(n):

      rf = r.copy()
      rb = r.copy()

      rf[j] = rf[j] + h
      rb[j] = rb[j] - h

      J[i,j] = (f[i](rf[0],rf[1]) - f[i](rb[0],rb[1]))/(2*h)

  return J
```

```python
def Newton(x0, f, tol=1e-16, max_iter=1000):
    error = 1
    it = 0
    x = np.copy(x0)

    for it in range(max_iter):
        vector = GetF2(f, x)
        jacobiana = GetJacobian2(f, x)

        inversa = np.linalg.inv(jacobiana)
        delta_x = -np.dot(inversa, vector)

        x = x + delta_x

        error = np.linalg.norm(delta_x)
        if error < tol:
            break

    return x, it
```

```python
#Gradiente

def Metric(G,r):
  return 0.5*np.linalg.norm(GetF2(G,r))**2

def Minimizer(G,r,lr=1e-4,epochs=int(1e3),error=1e-4):

  metric = 1
  it = 0

  M = np.array([])
  R = np.array([r])

  while metric > error and it < epochs:

    M = np.append(M,Metric(G,r))

    J = GetJacobian2(G,r).T
    Vector = GetF2(G,r)

    r = r - lr*np.dot(J,Vector)

    R = np.vstack((R,r))

    metric = Metric(G,r)

    it += 1

  return r,it

xsol, it = Minimizer(G1,x0)
xsol
```

####13. Jacobiano de cuarto orden. Usando el operador de derivada central de orden 4 es posible mejoirar la estimación del Jacobiano para funciones vectoriales.

a) Escriba una función que estime el Jacobiano con el operador derivada de orden 4.

```python
def GetJacobian4(f, r, h=1e-3):
    n = r.shape[0]
    J = np.zeros((n, n))

    for i in range(n):
        for j in range(n):
            rf_2h = r.copy()
            rf_h = r.copy()
            rb_h = r.copy()
            rb_2h = r.copy()

            # Perturbar el valor de r en la variable j
            rf_2h[j] += 2 * h
            rf_h[j] += h
            rb_h[j] -= h
            rb_2h[j] -= 2 * h

            # Aplicar la fórmula de diferencias finitas de cuarto orden
            J[i, j] = (-f[i](rf_2h[0], rf_2h[1], rf_2h[2]) + 8 * f[i](rf_h[0], rf_h[1], rf_h[2]) -
                       8 * f[i](rb_h[0], rb_h[1], rb_h[2]) + f[i](rb_2h[0], rb_2h[1], rb_2h[2])) / (12 * h)

    return J
```

```python
import numpy as np

def levi_civita(i, j, k):
    if (i, j, k) in [(1, 2, 3), (2, 3, 1), (3, 1, 2)]:
        return 1
    elif (i, j, k) in [(3, 2, 1), (1, 3, 2), (2, 1, 3)]:
        return -1
    else:
        return 0

J1 = np.array([[0, 0, 0],[0, 0, -1],[0, 1, 0]])

J2 = np.array([[0, 0, 1],[0, 0, 0],[-1, 0, 0]])

J3 = np.array([[0, -1, 0],[1, 0, 0],[0, 0, 0]])

J = [J1, J2, J3]

conmutadores = [[None for _ in range(3)] for _ in range(3)]

for i in range(3):
    for j in range(3):
        conmutador = np.dot(J[i], J[j]) - np.dot(J[j], J[i])
        conmutadores[i][j] = conmutador

verificacion = True

for i in range(3):
    for j in range(3):
        for k in range(3):
            epsilon = levi_civita(i+1, j+1, k+1)
            if epsilon != 0 and not np.allclose(conmutadores[i][j], epsilon * J[k]):
                verificacion = False
                break

verificacion
```
