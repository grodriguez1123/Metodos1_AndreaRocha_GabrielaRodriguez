d) (Python)  Para f(x)  =√tan(x)  estimar  la  derivada  central  de  orden (h2)  en  el intervalo [0.1,1.1] con h= 0.01

```python
def DerivadaCentral(f,x,h=0.01):
  return (f(x+h) - f(x-h))/(2*h)

DerivadaCentral(f,x)
```

```python
from os import error
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym


def Lagrange(x, X, i):
    L = 1
    for j in range(len(X)):
        if i != j:
            L *= (x - X[j]) / (X[i] - X[j])
    return L

def funcion(x):
    return np.sqrt(x)

X = [100, 121, 144]
Y = [funcion(100), funcion(121), funcion(144)]


x_vals = np.linspace(100, 144, 100)
y_vals = funcion(x_vals)

def lagrange_polynomial(
    x, X, Y):
    n = len(X)
    P = 0
    for i in range(n):
        P += Y[i] * Lagrange(x, X, i)
    return P

error = np.abs(funcion(114) - lagrange_polynomial(114, X, Y))
print (error)
```
